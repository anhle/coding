##930
###Approach 1: Prefix sum
Find some interval [i:j] ,i < j where sum[i:j] = target
The problem reduces to finding # of pairs (i, j) (i < j) such that prefix[j] - prefix[i-1] = target

###Approach 2: 3 pointer
For each j (in increasing order), let's maintain 4 variables:

* sum_lo : the sum of subarray [i_lo, j]
* sum_hi : the sum of subarray [i_hi, j]
* i_lo : the smallest i so that sum_lo <= S
* i_hi : the largest i so that sum_hi <= S
Then, (provided that sum_lo == S), the number of subarrays ending in j is i_hi - i_lo + 1.

```python
def numSubarraysWithSum(self, A, S):
    cnt = collections.Counter({0:1})
    prefixSums = res = 0
    for x in A:
        prefixSums +=x
        res += cnt[prefixSums-S]
        cnt[prefixSums] +=1
    return res

def numSubarraysWithSum(self, A, S):
    #Time: O(n), space O(1)
    i_lo,i_hi = 0,0
    sum_lo,sum_hi = 0,0
    ans = 0
    for j,x in enumerate(A):
        # maintain i_lo,sum_lo while the sum is too big, i_lo +=1
        sum_lo +=x
        while i_lo < j and sum_lo > S:
            sum_lo -= A[i_lo]
            i_lo += 1
        # maintain i_hi,sum_hi
        sum_hi +=x
        while i_hi < j and (sum_hi > S or sum_hi == S and not A[i_hi]):
            sum_hi -= A[i_hi]
            i_hi += 1

        if sum_lo == S:
            ans += i_hi-i_lo +1
    return ans
```

##678
Approach 1: Greedy
When checking whether the string is valid, we only cared about the "balance": the number of extra, open left brackets as we parsed through the string.    

Approach 2: dynamic programming
Let dp[i][j] be true if and only if the interval s[i], s[i+1], ..., s[j] can be made valid. Then dp[i][j] is true only if:

* s[i] is '*', and the interval s[i+1], s[i+2], ..., s[j] can be made valid;

* or, s[i] can be made to be '(', and there is some k in [i+1, j] such that s[k] can be made to be ')', plus the two intervals cut by s[k] (s[i+1: k] and s[k+1: j+1]) can be made valid;

Approach 3: backtrack
Approach 4: using stack

```python
def checkValidString(self, s):
    leftCnt,rightCnt = 0,0
    for i in range(len(s)):
        if s[i] == ')':
            leftCnt -=1
        else:
            leftCnt +=1
        if leftCnt < 0: return False

        if s[len(s)-i-1] == '(':
            rightCnt -=1
        else:
            rightCnt +=1
        if rightCnt < 0: return False
    return True
```

Approach: 2 pointer

```python
def middleNode(self, head):
    slow = fast = head
    while fast and fast.next:
        slow,fast = slow.next,fast.next.next

    return slow
```

##206
Approach 1: iteractive
While you are traversing the list, change the current node's next pointer to point to its previous element. Since a node does not have reference to its previous node, you must store its previous element beforehand
Approach 2: recursive
    

```python
def reverseList(self, head):
    cur, prev = head,None
    while cur:
        cur.next, cur, prev = prev, cur.next, cur
    return prev
```

##622
Design Circular Queue

```python
class Node:
    def __init__(self,value):
        self.val = value
        self.next = self.pre = None
class MyCircularQueue:

    def __init__(self, k: int):
        """
        Initialize your data structure here. Set the size of the queue to be k.
        """
        self.size  = k
        self.cursize = 0
        self.head = self.tail = Node(-1)
        self.head.next = self.tail
        self.tail.pre = self.head

    def enQueue(self, value: int) -> bool:
        """
        Insert an element into the circular queue. Return true if the operation is successful.
        """
        if self.cursize <self.size:
            node = Node(value)
            node.pre = self.tail.pre
            node.next = self.tail
            self.tail.pre = node
            node.pre.next = node
            self.cursize +=1
            return True
        return False
            

    def deQueue(self) -> bool:
        """
        Delete an element from the circular queue. Return true if the operation is successful.
        """
        if self.cursize > 0:
            remove_node = self.head.next
            self.head.next = remove_node.next
            remove_node.next.pre = self.head
            self.cursize -=1
            return True
        return False

    def Front(self) -> int:
        """
        Get the front item from the queue.
        """
        return self.head.next.val

    def Rear(self) -> int:
        """
        Get the last item from the queue.
        """
        return self.tail.pre.val

    def isEmpty(self) -> bool:
        """
        Checks whether the circular queue is empty or not.
        """
        return self.cursize == 0

    def isFull(self) -> bool:
        """
        Checks whether the circular queue is full or not.
        """
        return self.cursize == self.size
```

##199 
Binary Tree Right Side View
Approach 1: BFS:Traverse the tree level by level and add the last value of each level to the view. This is O(n).

```python
def rightSideView_BFS(self, root: TreeNode) -> List[int]:
    view = []
    level = [root] if root else []
    while level:
        view.append(level[-1].val) # get rightmost node
        # build next level
        curLevel, level = level,[]
        for i in curLevel:
            if i.left:
                level.append(i.left)
            if i.right:
                level.append(i.right)
    return view


def rightSideView_DFS(self, root: TreeNode) -> List[int]:
    #DFS-traverse the tree right-to-left, add values to the view whenever we first reach a new record depth. This is O(n).
    def dfs(node,level):
        if node:
            if level == len(view):
                view.append(node.val)
            dfs(node.right,level+1)
            dfs(node.left,level+1)

    view = []
    dfs(root,0)
    return view
```
